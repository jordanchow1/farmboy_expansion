```python
import pandas as pd

stores = pd.read_csv('farmboy_stores.csv')

# Extract FSA from the Address column (first 3 characters of postal code)
stores['FSA'] = stores['Address'].str.extract(r'([A-Z]\d[A-Z])')

census_data = pd.read_csv('ontario_fsa.csv')
# Merge the datasets
merged_data = stores.merge(census_data, on='FSA', how='left')

# Check the result
print(merged_data.head(10))
```

```python
merged_data.to_csv('merged_data.csv', index=False)
```
